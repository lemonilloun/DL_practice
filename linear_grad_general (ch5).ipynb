<b>#1 Градиентный спуск для нескольких входных значений<b>

```python
def neural_network(input, weights):
    pred = w_sum(input, weights)
    return pred

def w_sum(a, b):
    assert(len(a) == len(b))
    output = 0
    
    for i in range(len(a)):
        output+= a[i] * b[i]
    
    return output

def ele_mul(number, vector):
    output = [0] * 3
    
    assert(len(output) == len(vector))
    
    for i in range(len(output)):
        output[i] = number * vector[i]
    
    return output
```

```python
weights = [0.1, 0.2, -0.1]

toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

win_or_lose_binary = [1, 1, 0, 1]

true = win_or_lose_binary[0]
input = [toes[0], wlrec[0], nfans[0]] #вектор

pred = neural_network(input, weights) #скаляр

error = (pred - true) ** 2

delta = pred - true

weight_deltas = ele_mul(delta, input)
#print(weight_deltas)

alpha = 0.01

for i in range(len(weights)):
    weights[i] -= alpha * weight_deltas[i]
print(f'Weights: {weights}')
print(f'Weights Deltas: {weight_deltas}')

#print(neural_network(input, weights))
```

<b>#2 Градиентный спуск для нескольких входных значений в нескольких иттерацях<b>

<b>#3 Градиентный спуск - замораживание одного веса (влияние весов друг на друга)<b>

```python
def neural_network(input, weights):
    out = 0
    for i in range(len(input)):
        out += (input[i] * weights[i])
    return out

def ele_mul(scalar, vector):
    out = [0] * 3
    for i in range(len(out)):
        out[i] = vector[i] * scalar
    return out
```

```python
weights = [0.1, 0.2, -0.1]

toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

win_or_lose_binary = [1, 1, 0, 1]

alpha = 0.3

true = win_or_lose_binary[0]
input = [toes[0], wlrec[0], nfans[0]]
```

```python
for iter in range(3):
    
    pred = neural_network(input, weights)
    
    error = (pred-true) ** 2
    delta = pred - true
    
    weight_deltas = ele_mul(delta, input)
    weight_deltas[0] = 0
    
    print(f'Iteration: {iter}')
    print(f'Prediction: {pred}')
    print(f'Error: {error}')
    print(f'Delta: {delta}')
    print(f'Weights: {weights}')
    print(f'Delta Weights: \n{weight_deltas}')
    print("--------------------------" + "\n")
    
    for i in range(len(weights)):
        weights[i] -= alpha * weight_deltas[i]
```

<b>#5 Обучение метод градиентоного спуска с несколькими выходами и несколькими входами<b>

```python
def vect_mat_mul(vect, matrix):
    assert(len(vect) == len(matrix))
    output = [0] * 3
    
    for i in range(len(vect)):
        output[i] = w_sum(vect, matrix[i])
    return output

def neural_network(input, weights):
    pred = vect_mat_mul(input, weights)
    return pred

def outer_prod(vec_a, vec_b):
    out = np.zeros((len(vec_a), len(vec_b)))
    
    for i in range(len(vec_a)):
        for j in range(len(vec_b)):
            out[i][j] = vec_a[i] * vec_b[j]
    return out
```
